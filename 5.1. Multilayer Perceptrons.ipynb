```python
import torch
import matplotlib.pyplot as plt
```

#### ReLU

```python
x = torch.arange(-8.0, 8.0, 0.1, requires_grad=True)
y = torch.relu(x)  # y_i = ReLU(x_i) → Jacobian은 대각행렬
plt.figure(figsize=(5, 3))
plt.plot(x.detach(), y.detach())  # detach()하면 autograd에서 떨어져 나옴
plt.show()
```

```python
y.backward(gradient=torch.ones_like(y))  # v_T@J 이 값이 x.grad에 들어감 (VJP: Vector-Jacobian Product)
plt.figure(figsize=(5, 3))
plt.plot(x.detach(), x.grad)  # detach()하면 autograd에서 떨어져 나옴
plt.show()  # 관례적으로 0에서의 미분값은 0
```

#### Sigmoid

```python
y = torch.sigmoid(x)
plt.figure(figsize=(5, 3))
plt.plot(x.detach(), y.detach())  # detach()하면 autograd에서 떨어져 나옴
plt.show()
```

```python
x.grad.data.zero_()
y.backward(torch.ones_like(y), retain_graph=True)
plt.figure(figsize=(5, 3))
plt.plot(x.detach(), x.grad)  # detach()하면 autograd에서 떨어져 나옴
plt.show()
```

#### Hyperbloic Tangent

```python
y = torch.tanh(x)
plt.figure(figsize=(5, 3))
plt.plot(x.detach(), y.detach())  # detach()하면 autograd에서 떨어져 나옴
plt.show()
```

```python
x.grad.data.zero_()
y.backward(torch.ones_like(y), retain_graph=True)
plt.figure(figsize=(5, 3))
plt.plot(x.detach(), x.grad)  # detach()하면 autograd에서 떨어져 나옴
plt.show()
```
