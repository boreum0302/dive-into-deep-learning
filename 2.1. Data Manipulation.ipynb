```python
X = torch.arange(12).reshape(3, 4)                            # (3, 4)
Y = torch.tensor([[2, 1, 4, 3], [1, 2, 3, 4], [4, 3, 2, 1]])  # (3, 4)
```

```python
torch.cat([X, Y], dim=0)  # shape: (dim0, dim1, dim2, ...)
                          # dim0끼리 더하기 → (6, 4)
```

```python
torch.cat([X, Y], dim=1)
```

```python
X == Y  # elementwise
```

```python
X.sum()
```

```python
X.sum(axis=0)  # dim0를 없애기 → (3, 4) → (4,)
```

```python
X.sum(axis=0, keepdims=True)  # dim0를 합치기 → (3, 4) → (1, 4)
```

```python
a = torch.arange(3).reshape((3, 1))
b = torch.arange(2).reshape((1, 2))
a, b
```

```python
a + b  # broadcasting → (3, 2)
```

```python
before = id(Y)
Y = Y + X  # new Y 생성!!
id(Y) == before
```

```python
before = id(X)
X += Y  # 이미 있던 X에 Y를 더함!!
id(X) == before
```
